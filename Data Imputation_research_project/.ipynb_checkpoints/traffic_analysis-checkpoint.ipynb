```python
plt.figure(figsize=(18, 8))
plt.plot(df.Hour, df['VHT'])
plt.xlabel("Hour", fontsize=14)
plt.ylabel("VHT (Veh-Hours)", fontsize=14)
plt.xticks(np.arange(0, len(df.Hour)+1, 1), rotation=90, fontsize=12)
plt.yticks(fontsize=14)
plt.title("Recorded Vehcle Hours", fontsize=16)
plt.show()
```

```python
window = 24
window_start_index = 0
knn_neighbors = 5
N = len(df)
print('Data Length:',N)

a_list = list(range(1, window+1))
print(a_list)
```

```python
while window_start_index <= N - window:
    if window_start_index == 0:
        working_data = df[window_start_index:window_start_index+window]
        Before_imputation = working_data
        Before_imputation.iloc[3,1] = np.nan;
        Before_imputation.Hour=a_list
        #print(Before_imputation.isna().sum())
        print(Before_imputation)      
        #print(Before_imputation.Hour)
        # create an object for KNNImputer
        imputer = KNNImputer(n_neighbors=knn_neighbors)
        After_imputation = imputer.fit_transform(Before_imputation)
        # print dataset after performing the operation
        print("\n\nAfter performing imputation\n",After_imputation)
    window_start_index += 1
```
